## リストとインデックス
以下では、ut_dep という変数に['理1','理2','理3','文1','2', '文3'] という文字列の要素からなるリストが入っています。ここで、ut_dep [0] を評価して値を見てみましょう。同様に、ut_dep [1] を評価するとどうなるでしょう。

```python
ut_dep = ['理1', '理2', '理3', '文1', '文2', '文3']
```

```python
print(ut_dep[0])

print("文科：" + ut_dep[3] + "," + ut_dep[4] + "," + ut_dep[5])

# 逆から
print("文科：" + ut_dep[-3] + "," + ut_dep[-2] + "," + ut_dep[-1])
```

### リストの入れ子もできる


```python
ut_dep2 = [['理1', '理2','理3'], ['文1', '文2', '文3']]
# 前者のリストの３番目
print(ut_dep2[0][2])
```

### リストの値の変更

```python
ut_dep = ['理', '文']
ut_dep[0] = "science"

print(ut_dep)
```

### リストのスライス

```python
ut_dep = ['理1', '理2', '理3', '文1', '文2', '文3']

ut_dep_science = ut_dep[:3]
ut_dep_human = ut_dep[3:]

print(ut_dep_science)
print(ut_dep_human)
```

### リストの結合

```python
ut_dep = ut_dep_human + ut_dep_science

print(ut_dep)
print(ut_dep*2)
```

### リストとfor文

```python
ut_dep=['理1','理2','理3','文1','文2','文3']

for dep in ut_dep:
    print(dep)
```

--- 

### リストのインデックスを使って繰り返す場合は`range()`と`len()`を用いる

```python
for i in range(len(ut_dep)):
    print(i, ut_dep[i])
```

### リストとin演算子

要素がリストの中に含まれているのか判定

```python
'理1' in ut_dep
```

### index()`で何番目にその要素があるかインデックスを返す

```python
ut_dep.index('理3')
```
