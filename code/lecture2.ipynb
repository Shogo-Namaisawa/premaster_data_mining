### `apend()`と`insert()`によるリスト編集

これらのメソッドは元のリストを直接書き換える操作であり、新しいリストを戻り値として返すわけではありません。

```python  
ut_dep = ['理2', '理3', '文1', '文2']

# ありがちな間違い：戻り値を新しい変数に入れようとする
new_list = ut_dep.append('文3')

print(new_list) # 結果は ['理2', ..., '文3'] ではなく、None になる
print(ut_dep)   # 元のリスト自体は書き換わっている
``` 

````python
ut_dep=['理1','理2','理3','文1','文2','文3']
ut_dep.append('情コミ')
print(ut_dep)

# insertはどこに追加するかを指定しないとだめ
ut_dep=['理1','理2','理3','文1','文2','文3']
ut_dep.insert(0, '滋賀大')
print(ut_dep)
````

### `remove()`と`pop()`, `sort()`によるリスト編集

- remove()は渡された値をリストから削除
- del 文はリストから削除したい値のインデックスがわかっている時に使い、 remove() メソッドはリストから削除したい値がわかっている時に使います。
- pop() メソッドは、渡された値のインデックスの要素をリストから削除し、削除した値を返します。
- sort() メソッドを使うと、リストの要素（数や文字列）を並べ換えることができます

````python
ut_dep=['理1','理2','理3','文1','文2','文3']

ut_dep.remove('理1')
print(ut_dep)
````

````python
ut_dep=['理1','理2','理3','文1','文2','文3']

# 抜いて出力
print(ut_dep.pop(3))
print(ut_dep)
````

````python
num = [4,3,5,1,2]
print(num)

num.sort()
print(num)


# 文字列の並び替えは、ASCIIコード順で行う
words = ['desert','banana','carrot','apple']
print(words)

words.sort()
print(words)

# アルファベット順で行いたい場合は、キーワード引数keyにstr.lowerを指定
word = ['Desert','Banana','carrot','apple']
print(word)

word.sort(key=str.lower)
print(word)
````

### 辞書のkeys, values, itemsメソッド

````python
course = {'semester':'Spring', 'title':'data mining ', 'year':'2025', 'nuit':'2'}

# 左側
for k in course.keys():
    print(k)

# 右側
for k in course.values():
    print(k)

# 両方
for k in course.items():
    print(k)
````

### 辞書とin演算子（`get()`と`pop()`）


````python
course = {'semester':'Spring', 'title':'data mining ', 'year':'2025'}

# 第一引数がyearの値があればその値、なければ2017とする
print(course.get('year', 2017))
print(course.get('nuit', 1))

course = {'semester':'Spring', 'title':'data mining ', 'year':'2025', 'unit':'2'}

print(course.pop('unit'))
# popしたから抜けている
print(course)
````

### 練習
文中の単語を一つずつ出力して、その際に大文字にして

`Data mining is the process of discovering patterns in large data sets involving methods at the intersection of machine learning, statistics, and database systems.`

````python
sentence = 'Data mining is the process of discovering patterns in large data sets involving methods at the intersection of machine learning, statistics, and database systems.'

words = sentence.split(' ')

for word in words:
  word = word.rstrip(',')
  # 単語と文字数をセットで出力
  print(word.upper(), len(word))
````

### write()ファイルに書き込み
- open() 関数の第2引数に 'w' を渡すと書き込みモードでファイルを開きます。書き込みモードでは、既存のファイルの内容を上書きして書き直します。
- open() 関数の第2引数に 'a' を渡すと追記モードでファイルを開きます。追記モードでは、既存のファイルの終端に追加して書き込みます。
- 書き込みモードでも追記モードでも、もし引数で指定したファイルが存在しなければ、新たに空のファイルが引数で示したパスに作成されます。
- ファイルを読み書きした後は、 close() メソッドを呼び出して閉じます。


````python
myfile = open('data/greeting.txt', 'w')
myfile.write('hello\n')
myfile.close()
````

````python
myfile = open('data/greeting.txt','a')
myfile.write('Bonjour\n')
myfile.close()

myfile = open('data/greeting.txt')
a = myfile.read()
print(a)
myfile.close()
````

### 練習2
"data_mining.txt"を開いてファイル中のdataの出現回数をカウントせよ。大文字と小文字は区別しない

````python
dmfile = open("data/data_mining.txt")

content = dmfile.readlines()
freq = 0

for line in content:
  # rstripで文字列末尾の指定した文字を削除
  sentence = line.rstrip().rstrip('.').split(' ')
  
  for word in sentence:
    word=word.rstrip(',')
    if word.lower() == "data":
      freq=freq+1
print(freq)
````

### csv モジュール

````python
import csv
score = open('data/exam_score.csv')

# readerで行ごとに順番処理可能
score_reader = csv.reader(score)
# listで一行１リストで結果を返せる
score_data = list(score_reader)
print(score_data)

# 2行目の3列目
print(score_data[1][2])
````

### readerからのfor文

````python
score = open('data/exam_score.csv')
score_reader = csv.reader(score)

for row in score_reader:
    # 、行番号を取得するのに、 Reader オブジェクトの line_num 変数を使う
    print(str(score_reader.line_num)+' :'+str(row))
````

### Writer オブジェクトでデータをCSVファイルに書き込む

````python
score = open('data/my_score.csv', 'w')

# 書き込みモードに
score_writer = csv.writer(score)

# writerowで行ごとに書き込み
score_writer.writerow(['user', 'kokugo', 'sugaku'])
score_writer.writerow([1,50,50])
score_writer.writerow([2,30,30])
````
