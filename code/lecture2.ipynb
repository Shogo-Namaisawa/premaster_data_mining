## リストとインデックス
以下では、ut_dep という変数に['理1','理2','理3','文1','2', '文3'] という文字列の要素からなるリストが入っています。ここで、ut_dep [0] を評価して値を見てみましょう。同様に、ut_dep [1] を評価するとどうなるでしょう。

````python
ut_dep = ['理1', '理2', '理3', '文1', '文2', '文3']
````

````python
print(ut_dep[0])

print("文科：" + ut_dep[3] + "," + ut_dep[4] + "," + ut_dep[5])

# 逆から
print("文科：" + ut_dep[-3] + "," + ut_dep[-2] + "," + ut_dep[-1])
````

### リストの入れ子もできる


````python
ut_dep2 = [['理1', '理2','理3'], ['文1', '文2', '文3']]
# 前者のリストの３番目
print(ut_dep2[0][2])
````

### リストの値の変更

````python
ut_dep = ['理', '文']
ut_dep[0] = "science"

print(ut_dep)
````

### リストのスライス

````python
ut_dep = ['理1', '理2', '理3', '文1', '文2', '文3']

ut_dep_science = ut_dep[:3]
ut_dep_human = ut_dep[3:]

print(ut_dep_science)
print(ut_dep_human)
````

### リストの結合

````python
ut_dep = ut_dep_human + ut_dep_science

print(ut_dep)
print(ut_dep*2)
````

### リストとfor文

````python
ut_dep=['理1','理2','理3','文1','文2','文3']

for dep in ut_dep:
    print(dep)
````

### リストのインデックスを使って繰り返す場合は`range()`と`len()`を用いる

### `apend()`と`insert()`によるリスト編集

これらのメソッドは元のリストを直接書き換える操作であり、新しいリストを戻り値として返すわけではありません。

```python  
ut_dep = ['理2', '理3', '文1', '文2']

# ありがちな間違い：戻り値を新しい変数に入れようとする
new_list = ut_dep.append('文3')

print(new_list) # 結果は ['理2', ..., '文3'] ではなく、None になる
print(ut_dep)   # 元のリスト自体は書き換わっている
``` 

````python
ut_dep=['理1','理2','理3','文1','文2','文3']
ut_dep.append('情コミ')
print(ut_dep)

# insertはどこに追加するかを指定しないとだめ
ut_dep=['理1','理2','理3','文1','文2','文3']
ut_dep.insert(0, '滋賀大')
print(ut_dep)
````

### `remove()`と`pop()`, `sort()`によるリスト編集

- remove()は渡された値をリストから削除
- del 文はリストから削除したい値のインデックスがわかっている時に使い、 remove() メソッドはリストから削除したい値がわかっている時に使います。
- pop() メソッドは、渡された値のインデックスの要素をリストから削除し、削除した値を返します。
- sort() メソッドを使うと、リストの要素（数や文字列）を並べ換えることができます

````python
ut_dep=['理1','理2','理3','文1','文2','文3']

ut_dep.remove('理1')
print(ut_dep)
````

````python
ut_dep=['理1','理2','理3','文1','文2','文3']

# 抜いて出力
print(ut_dep.pop(3))
print(ut_dep)
````

````python
num = [4,3,5,1,2]
print(num)

num.sort()
print(num)


# 文字列の並び替えは、ASCIIコード順で行う
words = ['desert','banana','carrot','apple']
print(words)

words.sort()
print(words)

# アルファベット順で行いたい場合は、キーワード引数keyにstr.lowerを指定
word = ['Desert','Banana','carrot','apple']
print(word)

word.sort(key=str.lower)
print(word)
````
