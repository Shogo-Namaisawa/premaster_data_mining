### SeriesとDataFrame
SeriesとDataFrameは、pandasがデータを保持・操作するための基本構造であり、それぞれ一次元テンソルと二次元テンソルにラベル（名前やインデックス）を付与したようなデータ型。

- **Series**: 1列分のデータ（一次元）。インデックス（行ラベル）と値を持つ。
- **DataFrame**: 複数のSeriesを束ねた表形式データ（二次元）。行を管理するインデックスと、列を管理するカラム名を持つ。
- **インデックス**: 行を特定するためのラベル。デフォルトでは0から始まる連番だが、文字列や日付など任意のラベルを設定できる。
- **辞書との対応関係**:
  - Seriesは、**インデックス**をキー、**各データ**を値とした辞書とみなせる。
  - DataFrameは、**カラム名**をキー、**各列**を値とした辞書とみなせる。

````python
# リスト ->シリーズ
s1 = pd.Series([0,1,2,3,4,5])
print(s1)


# 辞書
#  ->シリーズ
s2 = pd.Series({0:'foo', 1:'boo', 2:'noo'})
print(s2)
````

````python
df1 = pd.DataFrame(
  [[1,2,3], [4,5,6],[7,8,9]], 
  index=[10,11,12], # 行のindex名指定
  columns=['a', 'i', 'u'] # 列のカラム名指定
)

print(df1)

df2 = pd.DataFrame(
  {
    'Initial':['B','M','W'],
    'Name':['foo','huu','noo']
  },
    columns=['Name', 'Initial']
)

print(df2)
````

# 課題１
中央値はデータの中心傾向を観察するための記述統計の指標です。同じくデータの中心傾向
の指標である平均値と比べると、データの外れ値の影響を受けにくという特徴があります。1
変数の 個のデータを考えた時、中央値を求めるにはまず、データの値の大きさの順に並び替
えます。次に、データの数が奇数個の場合は、大きさの順に並び替えた 番目の値、
データの数が偶数個の場合は、大きさの順に並び替えた 番目と の値の平均の値、
をそれぞれデータの中央値とします。
分散はデータのばらつきを観察するための記述統計の指標です。個々のデータがデータ全体
の平均からどらぐらいばらついているかを表します。以下の通り、個々のデータの値 とデ
ータ全体の平均の値 の差の2乗和をデータ数 で割ったものが分散になります。

0から100までの整数を要素とする任意の長さのリストを引数で入力として受け取り、以下を
要素とするリストを返す avg_med_var() 関数を完成させてください。

````python
def avg_med_var(input_list):
    # 初期化
    mean_value = 0
    med_value = 0
    var_value = 0
    results = [] # 結果のリスト

    input_list.sort() # リスト要素を昇順に並び替え
    mid_idx = len(input_list) // 2 # リスト中央のインデックス

    if len(input_list) % 2 ==0: # リストが偶数か奇数の中央値を計算
        med_value = (input_list[mid_idx -1] + input_list[mid_idx]) /2
    else:
        med_value = input_list[mid_idx]

    for i in range(len(input_list)):
        mse_value = (input_list[i] - mean_value)**2
    var_value += mse_value

    results.append(mean_value)
    results.append(med_value)
    results.append(var_value)
    return results
````

````python
print(avg_med_var([1,4,3,2,5,6,8,7,9,10]))
print(avg_med_var([1,4,3,2,5,6,8,7,9,10,11]))
````

### 各教科の得点がすべて平均点以上であるようなユーザの数を返す above_mean() 関数

ステップ1: ゴールを言語化する
「各教科の得点がすべて平均点以上のユーザの数を返す」

ステップ2: 必要な部品を洗い出す
- 各教科の平均点を求める → どうやって？
- 各ユーザが全教科で平均以上かを判定する → どうやって？
- 条件を満たすユーザの数を数える → どうやって？

ステップ3: pandasで部品を1つずつ試す

ステップ4: 関数にまとめる

````python
def above_mean(df):
    subjects = df.drop('user', axis=1)

    means = subjects.mean()
    # 各セルが平均以上かのブール表 → 全教科Trueの行を判定 → 数える
    # .all() は全ての要素がTrueかどうかを判定するメソッドです。
    count = (subjects >= means).all(axis=1).sum()

    return count
````
